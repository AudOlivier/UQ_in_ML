# Prior over a network's weights

This notebook aims at giving some insight on the choice of prior over weights in a neural network

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Model
from keras.layers import Input, Dense, Lambda
from keras.initializers import VarianceScaling
from UQ_in_ML.general_utils import *
import tensorflow as tf
from functools import partial
```

```python
# Function that defines a network
def create_nn(units_per_layer, activation, scale_input=False, bounds=None):
    # create model using Functional API
    inputs = Input(shape=(1,))
    if scale_input:
        x_ = Lambda(lambda x: (x - (bounds[1]+bounds[0])/2)/(bounds[1]-bounds[0]))(inputs)
    else:
        x_ = inputs
    for j, units in enumerate(units_per_layer):
        x_ = Dense(units, activation=activation)(x_)
    predictions = Dense(1)(x_)
    model = Model(inputs=inputs, outputs=predictions)
    #print(model.layers[1].kernel)
    #print(model.layers[1].bias)
    return model

# Function that samples the weights and assign them to the nn
def sample_and_set_weights(nn, prior):
    all_weights = []
    UQ_layers = [layer for layer in nn.layers if ('dense' in layer.get_config()['name'].lower())]
    for l, layer in enumerate(UQ_layers):
        prior_l = dict([(key, val[l]) for key, val in prior.items()])
        prior_l['type'] = prior['type']
        current_weights = layer.get_weights()
        kernel_shape, bias_shape = current_weights[0].shape, current_weights[1].shape
        new_kernel, new_bias = sample_weights_for_one_layer(kernel_shape, bias_shape, prior_l)
        all_weights.extend(list(new_kernel.reshape((-1,))))
        layer.set_weights([new_kernel, new_bias])
    return nn, all_weights

# Function to plot several posterior and weights
def create_posterior(units_per_layer=(10, 10, 10,), activation='tanh', 
                     prior={'type': 'gaussian', 'variance': 1},
                     n_plots = 50, ):
    prior = preprocess_prior(prior, n_layers=len(units_per_layer)+1)
    print('Prior type: '+prior['type'])
    tmp = ''.join(['{}, ']*len(units_per_layer))+'{}'
    if prior['type'].lower() == 'gaussian':
        print('Prior variance in each layer = '+tmp.format(*prior['variance']))
        print('Corresponding length scale (MC dropout) = '+tmp.format(*[1/np.sqrt(var) for var in prior['variance']]))
    elif prior['type'].lower() == 'gaussian_mixture':
        print('Prior variance_1 in each layer = '+tmp.format(*prior['variance_1']))
        print('Prior variance_2 in each layer = '+tmp.format(*prior['variance_2']))
        print('Prior proba_1 in each layer = '+tmp.format(*prior['proba_1']))
    fig, ax = plt.subplots(ncols=2, figsize=(16,5))
    x = np.linspace(-2, 2, 100).reshape((-1,1))
    # create network
    nn = create_nn(units_per_layer=units_per_layer, activation=activation)
    # sample weights and do plots
    all_weights = []
    for n in range(n_plots):
        nn, all_weights_ = sample_and_set_weights(nn, prior=prior)
        all_weights.extend(all_weights_)
        ax[0].plot(x, nn.predict(x), color='grey')
    ax[1].hist(all_weights, 40)
    ax[1].set_title('histogram of the weights W in the network'); ax[1].set_xlabel('weights')
    ax[0].set_title('samples from the prior'); ax[1].set_xlabel('weights')
    plt.show(fig)
```

## Gaussian prior

In the first set of plots, all layers share the same prior.

```python
stds = [0.10, 1., 10.]
for std in stds:
    create_posterior(units_per_layer=(20, 20, 20, 20), activation='tanh', 
                     prior = {'type': 'gaussian', 'variance': std ** 2},
                     n_plots = 25, )
```

## Effect of last layer: scales the output

```python
stds_last = [0.10, 1., 10.]
for std in stds:
    create_posterior(units_per_layer=(20, 20, 20, 20), activation='tanh', 
                     prior = {'type': 'gaussian', 'variance': [1., 1., 1., 1., std ** 2]},
                     n_plots = 25, )
```

## Glorot normal initializer, also called Xavier normal initializer.

This is a well known initializer for neural nets, where weights are drawn from a truncated normal distribution with std. dev. sqrt(2/(fan_in+fan_out)). The good thing about it is that the output is scaled to be between -1, 1 (circa).

```python
units_per_layer = (20, 20, 20, 20,)
variances = compute_glorot_normal_variance(input_dim=1, units_per_layer=units_per_layer, output_dim=1)
create_posterior(units_per_layer=units_per_layer, activation='tanh', 
                 prior = {'type': 'gaussian', 'variance': variances},
                 n_plots = 25, )
```

## Mixture of Gaussians

From BayesByBackprop paper, variance_2 is usually set << 1, variance_1 > variance_2

```python
prior = {'type': 'gaussian_mixture',
         'variance_1': 1.,
         'variance_2': 0.01,
         'proba_1': 0.5}
create_posterior(units_per_layer=(20, 20, 20, 20), activation='tanh', prior = prior, n_plots = 25, )
```

```python
prior = {'type': 'gaussian_mixture',
         'variance_1': 1.,
         'variance_2': compute_glorot_normal_variance(input_dim=1, units_per_layer=units_per_layer, output_dim=1),
         'proba_1': 0.5}
create_posterior(units_per_layer=(20, 20, 20, 20), activation='tanh', prior = prior, n_plots = 25, )
```

```python
# If you increase the varianc of the last layer you can allow for larger outputs :)
prior = {'type': 'gaussian_mixture',
         'variance_1': [1., 1., 1., 1., 100.],
         'variance_2': compute_glorot_normal_variance(input_dim=1, units_per_layer=units_per_layer, output_dim=1),
         'proba_1': 0.5}
create_posterior(units_per_layer=(20, 20, 20, 20), activation='tanh', prior = prior, n_plots = 25, )
```

## Choose a prior for the three examples considered in the notebooks

```python
def set_problem(problem, n_data):
    if problem == 'cubic':
        var_n = 9.
        f = partial(f_cubic, var_n=var_n)
        xn = np.random.uniform(low=-4.5, high=4.5, size=(n_data,1))
        bounds = (-6, 6)
    elif problem == 'sin_cos':
        var_n = 0.1**2
        f = partial(f_sin_cos, var_n=var_n)
        bounds = (0, 10)
        xn = np.random.uniform(low=bounds[0], high=bounds[1], size=(n_data,1))
    elif problem == 'last':
        var_n = 0.02**2
        f = partial(g_homoscedastic, var_n=var_n)
        xn = np.random.uniform(low=0., high=0.5, size=(n_data,1))
        bounds = (-0.2, 1.2)
    yn = f(xn, noisy=True)
    x_plot = np.linspace(bounds[0], bounds[1], 100).reshape((-1,1))
    y_plot = f(x_plot, noisy=False)
    return f, var_n, bounds, xn, yn, x_plot, y_plot

# Function to plot several posterior and weights
def plot_posterior(prior, ax, bounds, units_per_layer=(20, 20, 20, 20,), activation='tanh', n_plots = 50, ):
    prior = preprocess_prior(prior, n_layers=len(units_per_layer)+1)
    print('Prior type: '+prior['type'])
    tmp = ''.join(['{}, ']*len(units_per_layer))+'{}'
    if prior['type'].lower() == 'gaussian':
        print('Prior variance in each layer = '+tmp.format(*prior['variance']))
        print('Corresponding length scale (MC dropout) = '+tmp.format(*[1/np.sqrt(var) for var in prior['variance']]))
    elif prior['type'].lower() == 'gaussian_mixture':
        print('Prior variance_1 in each layer = '+tmp.format(*prior['variance_1']))
        print('Prior variance_2 in each layer = '+tmp.format(*prior['variance_2']))
        print('Prior proba_1 in each layer = '+tmp.format(*prior['proba_1']))
    x = np.linspace(bounds[0], bounds[1], 100).reshape((-1,1))
    # create network
    nn = create_nn(units_per_layer=units_per_layer, activation=activation, bounds=bounds, scale_input=True)
    # sample weights and do plots
    all_weights = []
    for n in range(n_plots):
        nn, all_weights_ = sample_and_set_weights(nn, prior=prior)
        ax.plot(x, nn.predict(x), color='grey')
    ax.set_title('samples from the '+prior['type']+' prior');
    
def create_plot(problem, n_data, prior, units_per_layer=(20, 20, 20, 20,), activation='tanh', n_plots = 50, ):
    fig, ax = plt.subplots(ncols=3, figsize=(16, 4))
    
    f, var_n, bounds, xn, yn, x_plot, y_plot = set_problem(problem=problem, n_data=n_data)
    ax[0].plot(x_plot, y_plot, color='green', label='true function')
    ax[0].plot(xn, yn, color='blue', marker='.', linestyle='none', label='data')
    ax[0].legend()
    
    plot_posterior(prior=prior, ax=ax[1], bounds=bounds, 
                   units_per_layer=units_per_layer, activation=activation, n_plots=n_plots, )
    plt.show()
```

```python
units_per_layer=(14, 14, 14, 14,)
prior = {'type': 'gaussian', 'variance': [2]*len(units_per_layer)+[0.02]}
create_plot(problem='sin_cos', n_data=20, prior=prior, 
            units_per_layer=units_per_layer, activation='tanh', n_plots = 25, )
```

```python
units_per_layer=(14, 14, 14, 14,)
prior = {'type': 'gaussian', 'variance': [0.5]*len(units_per_layer)+[1000]}
create_plot(problem='cubic', n_data=20, prior=prior, 
            units_per_layer=units_per_layer, activation='tanh', n_plots = 25, )
```
